# **Fraud Detection**

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
# Load the dataset
cc_df = pd.read_csv('/content/creditcard.csv')
```

**Data Exploration**

```python
# dataset information
cc_df.info()
```

```python
# distribution of legit and fraudulent transaction
cc_df['Class'].value_counts()
```

Dataset is unbalanced


*   0 - Normal transaction
*   1 - Fraudulent transaction



```python
# separating the data for analysis
legit = cc_df[cc_df.Class == 0]
fraud = cc_df[cc_df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare values of both transactions
cc_df.groupby('Class').mean()
```

**Under-sampling**

Build a sample dataset containing distribution of Normal and Fraudulent transaction


Number of Fraudulent transaction - 492

```python
legit_sample = legit.sample(n=492)
```

Concatenate two dataframes

```python
new_data = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
X = new_data.drop(columns='Class', axis=1)
Y = new_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training**

Logistic Regression

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression model with training dataset
model.fit(X_train, Y_train)
```

```python
# Create a pipeline with StandardScaler and Logistic Regression
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Scaling features
    ('logistic', LogisticRegression(max_iter=500))  # Increased max_iter to 500
])

# Train the model
pipeline.fit(X_train, Y_train)
```

```python
print(classification_report(Y_test, Y_test))
```

**Model Evaluation**

Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```

accuracy of 94.28% on the training data and 94.92% on the test data suggests that your logistic regression model is performing consistently across both sets, with no major signs of overfitting or underfitting.
