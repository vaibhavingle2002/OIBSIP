2. Detect Outliers using the IQR Method

```python
# Calculate Q1 and Q3
Q1 = cd_df['Price'].quantile(0.25)
Q3 = cd_df['Price'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = cd_df[(cd_df['Price'] < lower_bound) | (cd_df['Price'] > upper_bound)]
print(f"Number of outliers: {outliers.shape[0]}")
```

3. Handling Outliers

```python
# Remove outliers from the dataset
data_no_outliers = cd_df[(cd_df['Price'] >= lower_bound) & (cd_df['Price'] <= upper_bound)]
```
